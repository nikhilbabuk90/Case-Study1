# 3. Plot a bar graph of male vs female buyers participated in the sales.

- Plotting the Bar graph by taking the gender and count of the cars

```python
sns.set(font_scale=1.4)
Cars['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Cars Purchased", labelpad=14)
plt.title("Male vs Female Buyers", y=1.02);
```

- Insight
- Female are purchasing more cars that male and count is more than 5000
